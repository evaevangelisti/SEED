```python
import uuid

mappings = []

with open("output_evangelisti.jsonl") as f:
    for line in f:
        try:
            data = json.loads(line)
        except json.JSONDecodeError:
            continue

        lemma = data.get("input", {}).get("input", {}).get("lemma", "")
        pos_tag = data.get("input", {}).get("input", {}).get("pos", "")
        senses = data.get("input", {}).get("input", {}).get("senses", [])

        try:
            mapping = json.loads(data.get("response", {}))
        except (TypeError, json.JSONDecodeError):
            mapping = None

        mappings.append({
            "id": str(
                uuid.uuid5(
                    uuid.NAMESPACE_DNS,
                    f"{lemma}|{pos_tag}|{'|'.join(sense['definition'].lower() for sense in senses)}",
                )
            ),
            "mapping": mapping
        })

mappings[0]
```

```python
import json
import pandas as pd
import spacy
from tqdm import tqdm

nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

def validate_bold_all(sentence, bold_offsets, lemma):
    valid_spans = []
    
    for start, end in bold_offsets:
        text = sentence[start:end]
        
        # 1️⃣ match diretto
        if text.lower() == lemma.lower():
            lemmas = [text.lower()]  # consideriamo il lemma "trovato" come lista
            valid_spans.append((start, end, text, lemmas))
            continue
        
        # 2️⃣ fallback: lemmatizzazione
        doc = nlp(text)
        lemmas = [token.lemma_.lower() for token in doc]
        
        if lemma.lower() not in lemmas:
            return None  # basta uno sbagliato → tutto invalido
        
        valid_spans.append((start, end, text, lemmas))
    
    return valid_spans


def get_spans_from_doc(doc, sentence, bold_offsets, lemma):
    spans = []
    for token in doc:
        # 1️⃣ match diretto (token identico al lemma)
        if token.text.lower() == lemma.lower():
            spans.append((token.idx, token.idx + len(token.text), token.text, [token.text.lower()]))
        # 2️⃣ match tramite lemmatizzazione
        elif token.lemma_.lower() == lemma.lower():
            spans.append((token.idx, token.idx + len(token.text), token.text, [token.lemma_.lower()]))
    
    if spans:
        return "lemma", spans
    else:
        return "none", []
    
def jsonl_to_dataframe_smart(path, batch_size=1000, n_process=1):
    rows = []
    
    sentences_for_nlp = []
    metadata_for_nlp = []
    
    with open(path, "r", encoding="utf-8") as f:
        for line in tqdm(f, desc="Reading JSONL"):
            entry = json.loads(line)
            lemma = entry["lemma"]
            entry_id = entry["id"]
            pos = entry.get("pos")
            
            for sense in entry.get("senses", []):
                sense_order = sense.get("sense_order")
                definition = sense.get("definition")
                
                for sentence_obj in sense.get("sentences", []):
                    
                    sentence = sentence_obj["sentence"]
                    bold_offsets = sentence_obj.get("word_offsets")
                    
                    # 1. prova bold con lemma (solo sul pezzo)
                    if bold_offsets:
                        valid_spans = validate_bold_all(
                            sentence,
                            bold_offsets,
                            lemma
                        )
                        
                        if valid_spans is not None:
                            rows.append({
                                "id": entry_id,
                                "lemma": lemma,
                                "pos": pos,
                                "sense_order": sense_order,
                                "definition": definition,
                                "sentence": sentence,
                                "method": "bold_lemma",
                                "n_matches": len(valid_spans),
                                "spans": valid_spans
                            })
                            continue
                    
                    # 2. fallback → serve NLP globale
                    sentences_for_nlp.append(sentence)
                    metadata_for_nlp.append({
                        "id": entry_id,
                        "lemma": lemma,
                        "pos": pos,
                        "sense_order": sense_order,
                        "definition": definition,
                        "sentence": sentence,
                        "bold_offsets": bold_offsets
                    })
    
    # 3. NLP solo dove serve
    docs = nlp.pipe(
        sentences_for_nlp,
        batch_size=batch_size,
        n_process=n_process
    )
    
    for doc, meta in tqdm(zip(docs, metadata_for_nlp), total=len(metadata_for_nlp), desc="NLP fallback"):
        
        method, spans = get_spans_from_doc(
            doc,
            meta["sentence"],
            meta["bold_offsets"],
            meta["lemma"]
        )
        
        rows.append({
            **meta,
            "method": method,
            "n_matches": len(spans),
            "spans": spans
        })
    
    return pd.DataFrame(rows)
```

```python
import json

count_valid = 0
count_total = 0
count_invalid = 0
count_analysis = 0
count_empty = 0

with open("output_evangelisti.jsonl") as f:
    for line in f:
        record = json.loads(line)
        count_total +=1

        response = record.get("response")
        if not response:
            count_empty += 1
            continue
        
        try:
            json.loads(response)
            count_valid += 1
        except json.JSONDecodeError:
            if response.startswith("analysis"):
                count_analysis += 1
            else:
                count_invalid += 1
            pass

print(f"Parsabili: {count_valid} / {count_total}")
print(f"Analisi: {count_analysis}")
print(f"Invalidi: {count_invalid}")
print(f"Vuoti: {count_empty}")
```

```python
import json
import re

def is_valid_response(response, senses, translations):
    # parse se è stringa
    if isinstance(response, str):
        try:
            response = json.loads(response)
        except json.JSONDecodeError:
            return False

    # deve essere dict
    if not isinstance(response, dict):
        return False

    n_senses = len(senses)
    n_trans = len(translations)

    # 1. lunghezza corretta
    if len(response) != n_senses:
        return False

    # 2. chiavi: F1 ... Fn
    expected_keys = {f"F{i}" for i in range(1, n_senses + 1)}
    if set(response.keys()) != expected_keys:
        return False

    seen_S = set()

    # 3. controlla valori
    for value in response.values():
        if value is None:
            continue

        # deve essere tipo "S<number>"
        if not isinstance(value, str) or not re.fullmatch(r"S\d+", value):
            return False

        num = int(value[1:])

        # 4. deve stare dentro translations
        if num < 1 or num > n_trans:
            return False

        # 5. no duplicati
        if num in seen_S:
            return False
        seen_S.add(num)

    return True


count_valid = 0
count_total = 0
valid = []

with open("output_evangelisti.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)
        count_total += 1

        response = record.get("response")
        senses = record.get("input", {}).get("input", {}).get("senses", [])
        translations = record.get("input", {}).get("input", {}).get("translations", [])

        if is_valid_response(response, senses, translations):
            count_valid += 1
            valid.append({
                "id": record.get("input", {}).get("input", {}).get("id"),
                "mapping": json.loads(response)
            })

print(f"Validi: {count_valid} / {count_total}")
```

```python
import json
import random

rows = []

with open("output_evangelisti.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)

        response = record.get("response")
        if not response:
            continue

        try:
            mapping = json.loads(response)
        except:
            continue

        senses = record.get("input", {}).get("input", {}).get("senses", [])
        translations = record.get("input", {}).get("input", {}).get("translations", [])
        translation_keys = list(translations.keys())

        lemma = record.get("input", {}).get("input", {}).get("lemma", [])
        rec_id = record.get("input", {}).get("input", {}).get("id", [])

        # per ogni senso
        for i, sense in enumerate(senses, start=1):
            long_def = sense.get("definition")

            key = f"F{i}"
            mapped = mapping.get(key)

            if mapped is None:
                continue
            else:
                idx = int(mapped[1:]) - 1  # S1 -> 0
                if 0 <= idx < len(translation_keys):
                    short_def = translation_keys[idx]
                else:
                    continue

            rows.append({
                "id": rec_id,
                "lemma": lemma,
                "full_definition": long_def,
                "short_definition": short_def
            })

# 🔹 sampling finale
random.seed(42)
rows_sample = random.sample(rows, min(10000, len(rows)))

print(len(rows_sample))
print(rows_sample[:3])
```

```python
import json
from pathlib import Path

input_path = Path("data/raw/mappings.jsonl")
output_path = Path("mappings_fixed.jsonl")

with input_path.open(encoding="utf-8") as infile, output_path.open("w", encoding="utf-8") as outfile:
    for line in infile:
        try:
            entry = json.loads(line)
        except json.JSONDecodeError:
            continue

        try:
            entry["mapping"] = json.loads(entry.get("mapping"))
        except (TypeError, json.JSONDecodeError):
            entry["mapping"] = None

        outfile.write(json.dumps(entry, ensure_ascii=False) + "\n")
```

```python
import json
import random
import csv

random.seed(42)

rows = []

with open("data/processed/seed.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        entry = json.loads(line)
        lemma = entry.get("lemma")

        for sense in entry.get("senses", []):
            for s in sense.get("sentences", []):
                sentence = s["sentence"]
                offsets = s.get("word_offsets", [])

                if not offsets:
                    continue

                bold_sentence = sentence

                # inserisce ** partendo da destra
                for start, end in sorted(offsets, reverse=True):
                    bold_sentence = (
                        bold_sentence[:start]
                        + "**" + bold_sentence[start:end] + "**"
                        + bold_sentence[end:]
                    )

                rows.append((lemma, bold_sentence))

sample = random.sample(rows, 100)

with open("sample_sentences.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["lemma", "sentence"])
    writer.writerows(sample)
```

```python
import json
from collections import Counter

file_path = "data/processed/seed.jsonl"

ids = []
records = []

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)
        records.append(record)
        ids.append(record["id"])

# Conta quante volte appare ogni ID
id_counts = Counter(ids)

# Trova solo gli ID duplicati
duplicate_ids = {id_val for id_val, count in id_counts.items() if count > 1}

print(f"Numero di ID duplicati: {len(duplicate_ids)}\n")

# Stampa tutti i record che hanno un ID duplicato
for record in records:
    if record["id"] in duplicate_ids:
        print(record)
```
